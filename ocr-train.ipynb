```python
from PIL import Image, ImageDraw, ImageFont 
import os 
 
output_dir = 'synthetic_data' 
if not os.path.exists(output_dir): 
    os.makedirs(output_dir) 
 
 
fonts = [ 
    "fonts/gost_2.304_Bold.ttf", 
    "fonts/gost_2.304.ttf"
] 

for i, text in enumerate(texts):
    for font_path in fonts:
        if not os.path.exists(font_path):
            print(f"Шрифт не найден: {font_path}")
            continue
        try:
            font = ImageFont.truetype(font_path, 30)
            width, height = font.getsize(text)
            image = Image.new('RGB', (width + 10, height + 10), (255, 255, 255))
            draw = ImageDraw.Draw(image)
            draw.text((5, 5), text, font=font, fill=(0, 0, 0))
            font_name = os.path.splitext(os.path.basename(font_path))[0]
            output_file = os.path.join(output_dir, f'{font_name}_{i}.jpg')
            print("will save")
            image.save(output_file)
            print(f"Изображение сохранено: {output_file}")
            image.close()  # Закрытие изображения явно
        except Exception as e:
            print(f"Ошибка при обработке шрифта {font_path} с текстом '{text}': {e}")

print("Синтетические данные созданы и сохранены в директорию synthetic_data")
```

```python
import os
from trdg.generators import (
    GeneratorFromDict,
    GeneratorFromRandom,
    GeneratorFromStrings,
    GeneratorFromWikipedia,
)

# The generators use the same arguments as the CLI, only as parameters
generator = GeneratorFromStrings(
    texts,
    fonts=['fonts/gost_2.304_Bold.ttf', 'fonts/gost_2.304.ttf'],
    blur=1,
    random_blur=True
)
count = 0
output_dir = "synthetic_data"
for img, lbl in generator:
    output_file = os.path.join(output_dir, f'{count}.jpg')
    count += 1
    img.save(output_file) 
```

```python
from spellchecker import SpellChecker
import pickle
import pandas as pd
data = [
    "OptiDin",
]
def clean(word):
    return word.replace("(", "").replace(")", "")

dictionary = SpellChecker(language='ru', distance=1)

df = pd.read_csv('EKF спецификации.csv', sep=';')
print(df)

for name in df['Наименование']:
    data.append(clean(name))


def learn_spell(data):
    words = set()
    for name in data:
        for word in name.split():
            words.add(word)
    dictionary.word_frequency.load_words(words)
    print(words)
    with open('dictionary.pickle', 'wb') as f:
        pickle.dump(dictionary, f)
    print("Dictionary is ready")
    
    
learn_spell(data)
```
